## 3. SalePrice Analysis

### 3.1 Descriptive Statistics Summary

```python
SP = pd.concat([SalePrice_org, house_train.SalePrice], axis=1, keys=['SalePrice', 'SalePrice(log)'])
SP.describe()
```

### 3.2 Histogram and Distribution

```python
mean, std = norm.fit(SP['SalePrice'])
print( 'SalePrice(raw) \t\t : mean = {:.2f} and std = {:.2f}'.format(mean, std))
mean2, std2 = norm.fit(house_train['SalePrice'])
print( 'SalePrice(log-normal) \t : mean = {:.2f} and std = {:.2f}'.format(mean2, std2))
```

```python
SP['SalePrice'].describe()
```

```python
sns.distplot(SP['SalePrice'] , fit=norm);
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mean, std)], loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice(raw) distribution')

fig = plt.figure()
res = stats.probplot(SP['SalePrice'], plot=plt)
plt.show()
```

With pictures above we can see that the prices are skewed right and some outliers lies above ~500,000. We needt to get rid of the them to get a normal distribution of the independent variable (`SalePrice`).

```python
from sklearn.preprocessing import LabelEncoder

def scaleColumns(df, cols_to_scale):
   yy=LabelEncoder()
   for col in cols_to_scale:
       df[col] = pd.DataFrame(yy.fit_transform(pd.DataFrame(df[col])),columns=[col])
   return df
scaleColumns(train, ['GarageCond', 'KitchenQual'])
train['GarageCond']
```
